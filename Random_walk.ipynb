# Random Walk 

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.regression.linear_model import yule_walker
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_style('whitegrid')


%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = [13,4]
```

```python
steps = np.random.standard_normal(1000)
steps[0]=0
random_walk = np.cumsum(steps)
```

```python
random_walk[:10]
```

```python
plt.plot(random_walk)
plt.title("Simulated Random Walk")
plt.show()
```

```python
random_walk_acf_coef = plot_acf(random_walk)
```

```python
plot_acf(random_walk, lags=20);
```

```python
random_walk[:5]
```

```python
np.diff(random_walk[:5], n=1)
```

```python
random_walk_diff = np.diff(random_walk, n=1)
```

```python
# plt.figure(figsize=[10, 7.5]); # Set dimensions for figure
plt.plot(random_walk_diff)
plt.title('Noise')
plt.show()
```

```python
plot_acf(random_walk_diff, lags=20);
```

Significant peak at lag 0 after that everything is insignificant
